## Análisis exploratorio de datos (eda) inicial para cada conjunto de datos

Para cada uno de los cuatro conjuntos de datos, realizar un Análisis Exploratorio de Datos inicial. Esto incluye estadísticas descriptivas, visualizaciones y la identificación de valores faltantes y atípicos. Documentar los hallazgos en notebooks separados, con todas las explicaciones y comentarios en español.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el conjunto de datos de Iris
# Asumiendo que el archivo 'iris.csv' está disponible localmente
try:
    df_iris = pd.read_csv('iris.csv')
except FileNotFoundError:
    # Si no se encuentra localmente, intentar cargar desde una URL conocida si es posible
    print("Archivo 'iris.csv' no encontrado localmente. Intentando cargar desde URL...")
    url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
    column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
    df_iris = pd.read_csv(url, names=column_names)
    print("Conjunto de datos de Iris cargado desde URL.")


# 2. Mostrar las primeras filas
print("Primeras filas del conjunto de datos de Iris:")
display(df_iris.head())

# 3. Obtener resumen de estadísticas descriptivas
print("\nEstadísticas descriptivas del conjunto de datos de Iris:")
display(df_iris.describe())

# 4. Mostrar información sobre tipos de datos y valores no nulos
print("\nInformación del conjunto de datos de Iris:")
display(df_iris.info())

# 5. Identificar valores faltantes
print("\nValores faltantes por columna en el conjunto de datos de Iris:")
display(df_iris.isnull().sum())
```

```python
# 6. Generar visualizaciones
print("\nVisualizaciones para explorar la distribución y relaciones:")

# Histogramas para la distribución de las variables numéricas
df_iris.hist(figsize=(10, 8))
plt.suptitle("Histogramas de las Características de Iris", y=1.02)
plt.tight_layout()
plt.show()

# Diagramas de caja (Box plots) para identificar posibles atípicos
plt.figure(figsize=(12, 8))
df_iris.boxplot()
plt.title("Diagramas de Caja de las Características de Iris", fontsize=14)
plt.ylabel("Valor")
plt.show()

# Diagramas de dispersión (Scatter plots) para explorar relaciones entre pares de variables
# Usando seaborn para colorear por clase
sns.pairplot(df_iris, hue="class")
plt.suptitle("Diagramas de Dispersión de las Características de Iris por Clase", y=1.02)
plt.show()

# 7. Identificar posibles valores atípicos
print("\nObservaciones sobre posibles valores atípicos (basado en Box plots):")
print("- 'sepal_width' parece tener algunos valores atípicos, especialmente por encima de 4.0 y por debajo de 2.2.")
print("- Las otras características ('sepal_length', 'petal_length', 'petal_width') parecen tener distribuciones más simétricas con menos atípicos aparentes en los box plots generales.")
print("Nota: La identificación de atípicos puede depender del contexto y del método utilizado. Estos son solo observaciones iniciales basadas en la visualización.")

# 8. Documentar hallazgos (ya incluido en los comentarios y texto de las celdas)
print("\nDocumentación de hallazgos:")
print("- El conjunto de datos de Iris contiene 150 muestras y 5 columnas (4 características numéricas y 1 columna de clase).")
print("- No hay valores faltantes en ninguna columna.")
print("- Las estadísticas descriptivas muestran rangos y distribuciones variadas para las características numéricas.")
print("- Los histogramas y diagramas de dispersión muestran claras separaciones entre las clases, especialmente basadas en 'petal_length' y 'petal_width'.")
print("- Los diagramas de caja sugieren la presencia de posibles valores atípicos en 'sepal_width'.")
```

```python
# Iniciar EDA para el conjunto de datos de Precios de Casas de Boston

# Cargar el conjunto de datos de Precios de Casas de Boston
# Este conjunto de datos no está directamente disponible en pandas datasets y a menudo se carga de sklearn
# o se lee de un archivo. Asumiremos que el archivo 'boston_house_prices.csv' está disponible localmente
try:
    df_boston = pd.read_csv('boston_house_prices.csv')
    print("Conjunto de datos de Precios de Casas de Boston cargado desde archivo.")
except FileNotFoundError:
    print("Archivo 'boston_house_prices.csv' no encontrado localmente. Intentando cargar desde sklearn...")
    try:
        from sklearn.datasets import load_boston
        boston = load_boston()
        df_boston = pd.DataFrame(boston.data, columns=boston.feature_names)
        df_boston['PRICE'] = boston.target
        print("Conjunto de datos de Precios de Casas de Boston cargado desde sklearn.")
    except ImportError:
        print("La librería sklearn no está instalada o el conjunto de datos de Boston no está disponible en esta versión.")
        print("Por favor, asegúrese de tener sklearn instalado y que el conjunto de datos de Boston esté disponible.")
        # Si no se puede cargar, podemos terminar la tarea para señalar un problema.
        df_boston = None

if df_boston is not None:
    # 2. Mostrar las primeras filas
    print("\nPrimeras filas del conjunto de datos de Precios de Casas de Boston:")
    display(df_boston.head())

    # 3. Obtener resumen de estadísticas descriptivas
    print("\nEstadísticas descriptivas del conjunto de datos de Precios de Casas de Boston:")
    display(df_boston.describe())

    # 4. Mostrar información sobre tipos de datos y valores no nulos
    print("\nInformación del conjunto de datos de Precios de Casas de Boston:")
    display(df_boston.info())

    # 5. Identificar valores faltantes
    print("\nValores faltantes por columna en el conjunto de datos de Precios de Casas de Boston:")
    display(df_boston.isnull().sum())

    # 6. Generar visualizaciones
    print("\nVisualizaciones para explorar la distribución y relaciones en el conjunto de datos de Boston:")

    # Histogramas
    df_boston.hist(figsize=(15, 10))
    plt.suptitle("Histogramas de las Características de Precios de Casas de Boston", y=1.02)
    plt.tight_layout()
    plt.show()

    # Diagramas de caja
    plt.figure(figsize=(15, 10))
    df_boston.boxplot()
    plt.title("Diagramas de Caja de las Características de Precios de Casas de Boston", fontsize=14)
    plt.ylabel("Valor")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Diagramas de dispersión para algunas relaciones clave (ej: con el precio)
    print("\nDiagramas de dispersión mostrando la relación con el Precio:")
    fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 15))
    axes = axes.flatten()
    boston_features = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
    for i, col in enumerate(boston_features[:9]): # Mostrar los primeros 9 para no sobrecargar
        sns.scatterplot(x=col, y='PRICE', data=df_boston, ax=axes[i])
        axes[i].set_title(f'{col} vs Precio')
        axes[i].set_xlabel(col)
        axes[i].set_ylabel('Precio')
    plt.tight_layout()
    plt.show()


    # 7. Identificar posibles valores atípicos
    print("\nObservaciones sobre posibles valores atípicos (basado en Box plots):")
    print("- Varias características muestran una presencia significativa de valores atípicos (ej: CRIM, ZN, RM, B, LSTAT).")
    print("- 'CHAS' es una variable binaria y no presenta atípicos en el mismo sentido que las variables continuas.")
    print("- 'RAD', 'TAX', 'PTRATIO' también muestran algunas indicaciones de valores atípicos.")
    print("Nota: Es crucial investigar estos atípicos para determinar si son errores de datos o representan observaciones válidas pero inusuales.")

    # 8. Documentar hallazgos
    print("\nDocumentación de hallazgos para el conjunto de datos de Precios de Casas de Boston:")
    print("- El conjunto de datos contiene información sobre propiedades residenciales en áreas de Boston.")
    print("- Incluye 13 características predictoras y la variable objetivo 'PRICE'.")
    print("- No se encontraron valores faltantes en este conjunto de datos (si se cargó correctamente).")
    print("- Las estadísticas descriptivas y los histogramas revelan distribuciones variadas, algunas sesgadas.")
    print("- Los diagramas de dispersión muestran diferentes tipos de relaciones entre las características y el precio (lineales, no lineales, o ninguna clara).")
    print("- Los diagramas de caja indican la presencia de numerosos valores atípicos en varias características, lo cual es un desafío potencial para algunos modelos.")
else:
    print("\nNo se pudo cargar el conjunto de datos de Precios de Casas de Boston. No se pudo continuar con el EDA para este conjunto de datos.")
```

```python
# Continuar con EDA para el conjunto de datos de Clientes de Centro Comercial

# Cargar el conjunto de datos de Clientes de Centro Comercial
# Asumiendo que el archivo 'Mall_Customers.csv' está disponible localmente
try:
    df_mall = pd.read_csv('Mall_Customers.csv')
    print("Conjunto de datos de Clientes de Centro Comercial cargado desde archivo.")
except FileNotFoundError:
    print("Archivo 'Mall_Customers.csv' no encontrado localmente. No se puede continuar con el EDA para este conjunto de datos.")
    df_mall = None

if df_mall is not None:
    # Renombrar columnas para mayor claridad si es necesario (opcional, pero útil)
    df_mall.columns = ['CustomerID', 'Gender', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)']

    # 2. Mostrar las primeras filas
    print("\nPrimeras filas del conjunto de datos de Clientes de Centro Comercial:")
    display(df_mall.head())

    # 3. Obtener resumen de estadísticas descriptivas
    print("\nEstadísticas descriptivas del conjunto de datos de Clientes de Centro Comercial:")
    display(df_mall.describe())

    # 4. Mostrar información sobre tipos de datos y valores no nulos
    print("\nInformación del conjunto de datos de Clientes de Centro Comercial:")
    display(df_mall.info())

    # 5. Identificar valores faltantes
    print("\nValores faltantes por columna en el conjunto de datos de Clientes de Centro Comercial:")
    display(df_mall.isnull().sum())

    # 6. Generar visualizaciones
    print("\nVisualizaciones para explorar la distribución y relaciones en el conjunto de datos de Clientes de Centro Comercial:")

    # Distribución de variables clave (Edad, Ingresos Anuales, Puntuación de Gasto)
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    sns.histplot(df_mall['Age'], kde=True)
    plt.title('Distribución de Edad')
    plt.xlabel('Edad')
    plt.ylabel('Frecuencia')

    plt.subplot(1, 3, 2)
    sns.histplot(df_mall['Annual Income (k$)'], kde=True)
    plt.title('Distribución de Ingresos Anuales (k$)')
    plt.xlabel('Ingresos Anuales (k$)')
    plt.ylabel('Frecuencia')

    plt.subplot(1, 3, 3)
    sns.histplot(df_mall['Spending Score (1-100)'], kde=True)
    plt.title('Distribución de Puntuación de Gasto (1-100)')
    plt.xlabel('Puntuación de Gasto (1-100)')
    plt.ylabel('Frecuencia')

    plt.tight_layout()
    plt.show()

    # Diagrama de barras para el Género
    plt.figure(figsize=(6, 4))
    sns.countplot(x='Gender', data=df_mall)
    plt.title('Distribución por Género')
    plt.xlabel('Género')
    plt.ylabel('Conteo')
    plt.show()

    # Diagramas de dispersión para relaciones entre variables numéricas clave
    print("\nDiagramas de dispersión mostrando relaciones entre variables numéricas:")
    sns.pairplot(df_mall[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']])
    plt.suptitle("Diagramas de Dispersión entre Edad, Ingresos Anuales y Puntuación de Gasto", y=1.02)
    plt.show()

    # Relación entre Ingresos Anuales y Puntuación de Gasto, coloreado por Género
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Gender', data=df_mall)
    plt.title('Relación entre Ingresos Anuales y Puntuación de Gasto por Género')
    plt.xlabel('Ingresos Anuales (k$)')
    plt.ylabel('Puntuación de Gasto (1-100)')
    plt.show()


    # 7. Identificar posibles valores atípicos
    print("\nObservaciones sobre posibles valores atípicos (basado en visualizaciones):")
    print("- 'Annual Income (k$)' parece tener algunos valores atípicos en el extremo superior.")
    print("- 'Age' y 'Spending Score (1-100)' parecen tener distribuciones relativamente uniformes sin atípicos extremos aparentes en los histogramas.")
    print("Nota: Se podrían usar box plots para una identificación más formal de atípicos si fuera necesario.")

    # 8. Documentar hallazgos
    print("\nDocumentación de hallazgos para el conjunto de datos de Clientes de Centro Comercial:")
    print("- El conjunto de datos contiene información sobre clientes de un centro comercial, incluyendo ID, género, edad, ingresos anuales y puntuación de gasto.")
    print("- Contiene 200 entradas y 5 columnas.")
    print("- No se encontraron valores faltantes.")
    print("- La distribución de género es relativamente equilibrada.")
    print("- Las distribuciones de Edad y Puntuación de Gasto son algo uniformes, mientras que los Ingresos Anuales están más concentrados en rangos medios, con algunos valores más altos.")
    print("- Los diagramas de dispersión entre Ingresos Anuales y Puntuación de Gasto sugieren la presencia de grupos distintos de clientes, lo que es relevante para tareas de clustering.")
else:
    print("\nNo se pudo cargar el conjunto de datos de Clientes de Centro Comercial. No se pudo continuar con el EDA para este conjunto de datos.")
```

```python
# Continuar con EDA para el conjunto de datos de Precios de Acciones

# Cargar el conjunto de datos de Precios de Acciones
# Este conjunto de datos puede variar mucho dependiendo de la fuente.
# Asumiremos que tenemos un archivo CSV genérico llamado 'stock_prices.csv' con columnas como 'Date', 'Open', 'High', 'Low', 'Close', 'Volume'.
try:
    df_stock = pd.read_csv('stock_prices.csv')
    print("Conjunto de datos de Precios de Acciones cargado desde archivo.")
except FileNotFoundError:
    print("Archivo 'stock_prices.csv' no encontrado localmente. No se puede continuar con el EDA para este conjunto de datos.")
    df_stock = None

if df_stock is not None:
    # Convertir la columna 'Date' a tipo datetime si existe
    if 'Date' in df_stock.columns:
        df_stock['Date'] = pd.to_datetime(df_stock['Date'])
        # Establecer 'Date' como índice puede ser útil para series temporales
        df_stock.set_index('Date', inplace=True)

    # 2. Mostrar las primeras filas
    print("\nPrimeras filas del conjunto de datos de Precios de Acciones:")
    display(df_stock.head())

    # 3. Obtener resumen de estadísticas descriptivas
    print("\nEstadísticas descriptivas del conjunto de datos de Precios de Acciones:")
    display(df_stock.describe())

    # 4. Mostrar información sobre tipos de datos y valores no nulos
    print("\nInformación del conjunto de datos de Precios de Acciones:")
    display(df_stock.info())

    # 5. Identificar valores faltantes
    print("\nValores faltantes por columna en el conjunto de datos de Precios de Acciones:")
    display(df_stock.isnull().sum())

    # 6. Generar visualizaciones
    print("\nVisualizaciones para explorar las tendencias en el conjunto de datos de Precios de Acciones:")

    # Trazar las series de tiempo de los precios (Open, High, Low, Close)
    if all(col in df_stock.columns for col in ['Open', 'High', 'Low', 'Close']):
        df_stock[['Open', 'High', 'Low', 'Close']].plot(figsize=(15, 7))
        plt.title('Tendencia de Precios de Acciones a lo largo del Tiempo')
        plt.xlabel('Fecha')
        plt.ylabel('Precio')
        plt.show()

    # Trazar el volumen de negociación a lo largo del tiempo
    if 'Volume' in df_stock.columns:
        df_stock['Volume'].plot(figsize=(15, 5), color='grey')
        plt.title('Volumen de Negociación a lo largo del Tiempo')
        plt.xlabel('Fecha')
        plt.ylabel('Volumen')
        plt.show()

    # Histogramas para la distribución de los cambios diarios (si es posible calcularlos)
    if 'Close' in df_stock.columns:
        df_stock['Daily_Change'] = df_stock['Close'].pct_change().dropna()
        if not df_stock['Daily_Change'].empty:
            plt.figure(figsize=(8, 5))
            sns.histplot(df_stock['Daily_Change'], bins=50, kde=True)
            plt.title('Distribución del Cambio Diario de Precio')
            plt.xlabel('Cambio Diario')
            plt.ylabel('Frecuencia')
            plt.show()
        else:
             print("\nNo hay suficientes datos para calcular el cambio diario.")
    else:
        print("\nColumna 'Close' no encontrada para calcular el cambio diario.")


    # 7. Identificar posibles valores atípicos
    print("\nObservaciones sobre posibles valores atípicos:")
    print("- En series temporales, los 'atípicos' pueden ser picos o caídas inusuales en los precios o el volumen.")
    print("- Las visualizaciones de series de tiempo (gráficos de línea) son la mejor manera de identificar estos eventos inusuales visualmente.")
    print("- Los box plots también podrían usarse en las características numéricas si se consideran las fechas como puntos de datos independientes, pero es menos común para series temporales.")
    print("Nota: Los eventos atípicos en los precios de acciones a menudo corresponden a noticias importantes o eventos del mercado y no necesariamente son 'errores'.")


    # 8. Documentar hallazgos
    print("\nDocumentación de hallazgos para el conjunto de datos de Precios de Acciones:")
    print("- El conjunto de datos contiene datos históricos de precios y volumen de una acción (o varias).")
    print("- La granularidad de los datos es diaria (asumiendo el formato común).")
    print("- Se verificó la presencia de valores faltantes.")
    print("- Las visualizaciones de series de tiempo muestran las tendencias generales de precios y volumen.")
    print("- Se pueden observar picos o caídas inusuales que podrían considerarse atípicos en el contexto de una serie temporal.")
    print("- Este tipo de datos es adecuado para análisis de series temporales y tareas de predicción.")

else:
    print("\nNo se pudo cargar el conjunto de datos de Precios de Acciones. No se pudo completar el EDA para este conjunto de datos.")
```
